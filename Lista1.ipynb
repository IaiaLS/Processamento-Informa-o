Problema 1: Calcular a área de um retângulo.


```python
print("Calcula a área de um retângulo\n")
altura = float(input("Insira a altura: "))
base = float(input("Insira a base: "))
area = base*altura
print(f"Área do retângulo {area:0.2f}")
```

Problema 2: Converter graus Celsius para Fahrenheit

```python
print("Converter graus celsius para Fahrenheit\n")
celsius = float(input("Insira a temperatura em graus Celsius: "))
fahrenheit = (celsius*(9/5)) +32
print(f"Em Fahrenheit a temperatura é {fahrenheit:0.2f} °F")
```

Problema 3: Calcular o perímetro de um círculo

```python
import math
print("Calcula o perímetro de um círculo\n")
raio = float(input("Insira o raio: "))
perimetro = 2*raio*math.pi
print(f"O perímetro do círculo é {perimetro:0.2f}")
```

Problema 4: Calcular a média de três números

```python
print("Calcula a média de 3 números\n")
n1 = float(input("Insira o n1: "))
n2 = float(input("Insira o n2: "))
n3 = float(input("Insira o n3: "))
media = (n1+n2+n3)/3
print(f"A média é {media:0.2f}")
```

Problema 5: Calcular o volume de um cubo.


```python
print("Calcular Volume de um cubo\n")
aresta = float(input("Insira o valor da aresta: "))
volume = aresta*aresta*aresta
print(f"O volume do cubo é {volume:0.2f}")
```

Problema 6: Converter segundos em minutos.


```python
print("Converte segundos em minutos\n")
segundos = float(input("Insira os segundos: "))
minutos = segundos/60
print(f"{minutos:0.2f} minutos")
```

Problema 7: Calcular o valor do desconto em um produto a partir de seu valor e o
percentual de desconto.

```python
valor = float(input("Insira o valor do produto: "))
per_des= float(input("Insira o valor do percentual de desconto: "))
desconto = valor*(per_des/100)
print(f"O valor do desconto é {desconto:0.2f}")
```

Problema 8: Calcular o tempo gasto em uma viagem (em horas) a partir da distância e
da velocidade média.


```python
distancia = float(input("Insira a distância (km): "))
velocidade = float(input("Insira a velocidade média (em km/h): "))
tempo = distancia/velocidade
print(f"O tempo gasto foi de {tempo:0.2f} horas.")
```

Problema 9: Calcular o montante final após aplicar uma taxa de juros simples a um
empréstimo.


```python
c = float(input("Valor a ser emprestado: "))
i = float(input("Taxa mensal(%): "))
t = int(input("Tempo (em meses): "))
j = (c*i*t)/100
m = c+j
print(f"O montante final é {m:0.2f}")
```

Problema 10: Trocar o valor de duas variáveis
